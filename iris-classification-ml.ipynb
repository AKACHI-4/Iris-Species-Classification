```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.metrics import classification_report
from sklearn.model_selection import KFold, cross_val_score
```

```python
names = ['sepal_length','sepal_width','petal_length','petal_width','class']
iris_data = pd.read_csv('/kaggle/input/iris-flower-dataset/IRIS.csv', names=names)
```

```python
plt.scatter(iris_data['sepal_length'], iris_data['petal_length'])

plt.xlabel('Sepal length')
plt.ylabel('petal length')
plt.title('Sepal length vs petal length')

plt.show()
```

```python
iris_data['class'].value_counts().plot(kind='bar')

plt.xlabel('Species')
plt.ylabel('Number of samples')
plt.title('Number of samples by species')

plt.show()
```

```python
plt.plot(iris_data[iris_data['class']=='Iris-setosa']['sepal_length'], label='Iris-setosa')
plt.plot(iris_data[iris_data['class']=='Iris-versicolor']['sepal_length'], label='Iris-versicolor')
plt.plot(iris_data[iris_data['class']=='Iris-virginica']['sepal_length'], label='Iris-virginica')

plt.xlabel('Sample index')
plt.ylabel('Sepal length')
plt.title('Sepal length by species')

plt.legend()

plt.show()
```
